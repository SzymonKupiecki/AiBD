```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

Zad 3

```python
def f(x):
    return x*x + 5

def print_f(range, title, xlabel, ylabel, legend):
    plt.plot(range, f(range))
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.legend(legend)
    plt.show()

print_f(np.linspace(-1,1), "<-1,1>", "x", "y", ["function"])
print_f(np.linspace(-6,6), "<-6,6>", "x", "y", ["function"])
print_f(np.linspace(0,5), "<0,5>", "x", "y", ["function"])
```
